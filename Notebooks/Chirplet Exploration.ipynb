```python
from birdsong.datasets.tools.spectrograms import mel_s, chirp_s, stft_s
from birdsong.datasets.tools.io import load_audio
import matplotlib.pyplot as plt
```

```python
path = 'storage/top10ger/159521.mp3'
audio, sr = load_audio(path)
```

```python
len(audio)/sr #in seconds
```

```python
sr
```

```python
# a 5 second slice:
slice_length = sr * 5

slice_ = audio[3000:3000+slice_length]
```

```python
len(slice_)/sr
```

```python
chirp = chirp_s(slice_)

chirp.shape

plt.figure(figsize=(20, 20))
plt.imshow(chirp)
```

```python
mel = mel_s(slice_)
plt.figure(figsize=(20, 20))
plt.imshow(mel)
```

```python
stft = stft_s(slice_)
plt.figure(figsize=(20, 20))
plt.imshow(stft)
```
